```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import time
from scipy import stats
```

```python
from scipy.stats import linregress

data = pd.read_csv("dog_rates_tweets.csv")
data.set_index(['id','created_at','text'])
```

```python
data_scores = data[data['text'].str.contains('([0-9]/10)')==True]
data_scores['score'] = data['text'].str.extract(r'([0-9]/10)')
data_scores = data_scores.dropna(subset=['score'])    # Remove entries with NaN
data_scores['score'] = data_scores['score'].apply(pd.eval)
data_scores = data_scores.sort_values(by=['score'])
```

```python
data_scores['created_at'] = pd.to_datetime(data_scores['created_at'])
data_scores['timestamp'] = data_scores['created_at'].values.astype(np.timedelta64) / np.timedelta64(1, 'ms')
```

```python
x = data_scores['created_at']
x_ts = data_scores['timestamp']
y = data_scores['score']

res = linregress(x_ts, y)
data_scores['prediction'] = (res.intercept + res.slope*x_ts)
```

```python
data_scores
```

```python
res.slope, res.intercept
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=25)
plt.scatter(x, y, label='data', alpha=0.5)
plt.plot(x, res.intercept + res.slope*x_ts, 'r', linewidth=3, label='fitted line')
plt.title("Pup Inflation")
plt.ylabel("Scores")
plt.xlabel("Date Created")
plt.legend()
plt.show()
```

```python
data_scores['predict'] = res.slope*data_scores['timestamp'] + res.intercept
data_scores['residual'] = data_scores['score'] - data_scores['predict']
```

```python
plt.figure(figsize=(8,5))
plt.hist(data_scores['residual'])
plt.title("Residuals of Pup Inflation")
plt.xlabel("Residual")
plt.ylabel("Frequency")
```
